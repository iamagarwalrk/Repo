## Matplotlib Library

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
## Basic Graph
x=np.array([1, 2, 3, 4, 5])
y=np.array([2, 4, 6, 8, 10])

plt.figure(figsize=(5, 3), dpi=300) ## size and dpi control

## plot line-1 chart 
plt.plot(x, y, label="2x", color='y', linestyle="--", linewidth=2, marker=".", markeredgecolor='r') 

## plot 2 line

#x2=np.arrange(0, 4.5, 0.5) # not working why
plt.plot(x[:3], x[:3]**2, "r", label="x^2", marker="^") # with half line 1
plt.plot(x[2:], x[2:]**2, "r^--") # "linecolor, marker, linestyle" with half line 2

plt.title("Test Graph", fontdict={'fontname':"Comic Sans MS", "fontsize":25})

plt.xlabel("Number", fontdict={"fontname":"Times New Roman", "fontsize":15})
plt.ylabel("Number * 2", fontdict={"fontname":"Times New Roman", "fontsize":10})

plt.xticks([1, 2, 3, 4, 5]) #no fontdict method
plt.yticks([2, 4, 6, 8, 10, 20]) # automatically resize the graph


plt.legend()

plt.savefig("TestGraph.png", dpi=200)

plt.show()
```

```python
## Bar Chart

labels=['A', 'B', 'C']
values=[1,4,2]

bars=plt.bar(labels, values, color='blue',edgecolor='black') #its barh for horizontal

# method 1

"""
bars[0].set_hatch('/')
bars[2].set_hatch('.')
"""
# method 2

patterns=['/', ' ', '.']
for bar in bars:
    bar.set_hatch(patterns.pop(0)) ## pop from start

plt.show()
```

```python
## Histogram

fifa=pd.read_csv("FIFA.csv")

fifa.head()
fifa.shape
```

```python
bins=list(map(int, range(30, 110, 10)))

plt.hist(fifa.Overall, bins=bins, color="#8F00FF")
plt.title("Skill level distribution- FIFA")
plt.xlabel("Skill Level")
plt.ylabel("No. of players")
plt.xticks(bins)
plt.show()
```

```python
fifa.columns
```

```python
fifa[["Unnamed: 0", "Name", "Preferred Foot"]].head()
fifa["Preferred Foot"].unique
```

```python
## Pie Chart

left=fifa.loc[fifa["Preferred Foot"] == "Left"].count()[0] # to give no of all entry for true "Left"
right=fifa[fifa["Preferred Foot"]=="Right"].count()[0]
right+left
```

```python
plt.pie([left, right], labels=["left", "right"], colors=["red", "green"], autopct='%.2f %%') 
# for %% for % sign
plt.title("Prefereed Foot")
plt.show()
```

```python
## Pie Chart 2

fifa.Weight=[int(x.strip('lbs')) if type(x) == str else x for x in fifa.Weight] ## need to understand
fifa.Weight.head()

plt.style.use('ggplot')

## classification
light=fifa[fifa.Weight<125].count()[0]
light_medium=fifa[(fifa.Weight>=125) & (fifa.Weight<150)].count()[0] ## cant use and inplace of & why
medium=fifa[(fifa.Weight>=150) & (fifa.Weight<175)].count()[0]
heavy_medium=fifa[(fifa.Weight>=175) & (fifa.Weight<200)].count()[0]
heavy=fifa[fifa.Weight>=200].count()[0]

plt.pie([light, light_medium, medium, heavy_medium, heavy], labels=["light", "light medium", "medium", "heavy medium", "heavy"], autopct='%.2f %%', pctdistance=0.8, explode=[.4, 1, 0, 0, .4])
plt.show()
```

```python
## Box Whisker Chart

plt.style.use("default")

barcelona=fifa[fifa.Club=="FC Barcelona"]["Overall"]
madrid=fifa[fifa.Club=="Real Madrid"]["Overall"]

plt.figure(figsize=(5, 7))
plt.title('Professional Soccer Team Compare')

boxes=plt.boxplot([barcelona, madrid], labels=["barcelona", "madrid"])
plt.ylabel("FIFA overall Rating")


plt.show()
```
